```python
from Env import data
from pandas.plotting import autocorrelation_plot
import matplotlib.pyplot as plt
from arch import arch_model
```

```python
d = data()
```

```python
d.Returns_Visualization('spy',d.DataETFsReturns)
d.Returns_Visualization('spy',d.DataETFsAbsReturns)
```

```python
autocorrelation_plot(d.DataETFsReturns['spy'])
plt.title('Autocorrelation Function for SPY returns')
plt.xlabel('Lag')
plt.ylabel('Autocorrelation')
plt.show()

autocorrelation_plot(d.DataETFsAbsReturns['spy'])
plt.title('Autocorrelation Function for SPY absolute returns')
plt.xlabel('Lag')
plt.ylabel('Autocorrelation')
plt.show()
```

```python
returns = d.DataETFsReturns['spy']
model = arch_model(returns, mean='constant', vol='GARCH', p=1, q=1, dist='normal')
results = model.fit()
```

```python
print(results.summary())
```
